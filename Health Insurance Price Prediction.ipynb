##### Numerical Feature : Age, Children, Charges <br> Categorical Feature : Sex, Smoker, Region

##### Converting All Categorical Features Value into Numerical Value

```python
Clean_Data = {'sex': {'male' : 0 , 'female' : 1} ,
                 'smoker': {'no': 0 , 'yes' : 1},
                   'region' : {'northwest':0, 'northeast':1,'southeast':2,'southwest':3}
               }
New_Data = Data.copy()
New_Data.replace(Clean_Data, inplace=True)
```

```python
New_Data.describe()
```

##### Analyzing the Distribution for All Features

##### Analyzing Feature Distributions Across the Target Variable

##### Charges Distribution by Age

```python
plt.figure(figsize=(12,6))
plt.title('Age vs Charge')
sns.barplot(x='age', y='charges', data=New_Data, palette='husl')
plt.savefig('./Images/AgevsCharges')
plt.show()
```

##### Charges Distribution by Sex

```python
plt.figure(figsize=(5,5))
plt.title('Sex vs Charges')
sns.boxplot(x='sex', y='charges', data=New_Data, palette='Set2')
plt.savefig('./Images/SexvsCharges')
plt.show()
```

##### Charges Distribution by BMI

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='bmi', y='charges',  hue='sex', data=New_Data, palette='Reds')
plt.title('BMI VS Charge')
plt.savefig('./Images/BMIvsCharges')
plt.show()
```

##### Charges Distribution by Number of Children

```python
plt.figure(figsize=(8,6))
plt.title('Children vs Charges')
sns.boxplot(x='children', y='charges', data=New_Data, palette='Set2')
plt.savefig('./Images/ChildrenvsCharges')
plt.show()
```

##### Charges Distribution by Smoking Status

```python
plt.figure(figsize=(8,6))
plt.title('Smoker vs Charge')
sns.boxplot(x='smoker', y='charges', data=New_Data, palette='Set2', hue='sex')
plt.savefig('./Images/SmokervsCharges')
plt.show()
```

##### Charges Distribution by Region

```python
plt.figure(figsize=(8,6))
plt.title('Region vs Charge')
sns.boxplot(x='region', y='charges', data=New_Data, palette='Set2')
plt.savefig('./Images/RegionvsCharges')
plt.show()
```
